```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Admin\Downloads\archive (2).zip")

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

df.columns = df.columns.str.strip()

df.dropna(subset=['Time', 'Weather', 'Road_Condition'], inplace=True)

df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

df['Hour'] = df['Time'].dt.hour

plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=df, palette='viridis')
plt.title("Accidents by Time of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='Weather',
    data=df,
    order=df['Weather'].value_counts().index,
    palette='coolwarm'
)
plt.title("Accidents by Weather Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='Road_Condition',
    data=df,
    order=df['Road_Condition'].value_counts().index,
    palette='Set2'
)
plt.title("Accidents by Road Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Condition")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='Location',
    data=df,
    order=df['Location'].value_counts().head(10).index,
    palette='magma'
)
plt.title("Top 10 Accident Hotspots")
plt.xlabel("Number of Accidents")
plt.ylabel("Location")
plt.show()

print("\nConclusion:")
print("- Most accidents occur during peak traffic hours.")
print("- Weather conditions like rain and fog increase accident frequency.")
print("- Poor road conditions contribute significantly to accidents.")
print("- Certain locations are identified as accident hotspots.")
```
