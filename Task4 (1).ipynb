```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Admin\Downloads\archive (2).zip")

print(df.head())
print(df.info())
print(df.describe())

df.columns = df.columns.str.strip()

df.dropna(subset=['Time of Day', 'Weather Conditions', 'Road Condition'], inplace=True)

plt.figure(figsize=(10, 6))
sns.countplot(
    x='Time of Day',
    data=df,
    order=df['Time of Day'].value_counts().index,
    palette='viridis'
)
plt.title("Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='Weather Conditions',
    data=df,
    order=df['Weather Conditions'].value_counts().index,
    palette='coolwarm'
)
plt.title("Accidents by Weather Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Conditions")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='Road Condition',
    data=df,
    order=df['Road Condition'].value_counts().index,
    palette='Set2'
)
plt.title("Accidents by Road Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Condition")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    y='City Name',
    data=df,
    order=df['City Name'].value_counts().head(10).index,
    palette='magma'
)
plt.title("Top 10 Accident Hotspots (Cities)")
plt.xlabel("Number of Accidents")
plt.ylabel("City Name")
plt.show()

print("\nConclusion:")
print("- Accidents are influenced by specific times of the day.")
print("- Weather conditions such as rain and fog contribute to higher accident rates.")
print("- Poor road conditions increase accident frequency.")
print("- Certain cities are identified as major accident hotspots.")
```
